# Enunciado

Implemente um programa em C/C++ ou Python capaz de controlar outros processos. Esse processo recebe do teclado o endereço de um programa a ser executado e cria um processo filho para executar o programa. O processo controlador aguarda os seguintes comandos do teclado: s (suspende a execução do processo filho), c (continua a execução do processo filho) e t (termina a execução do processo filho). Caso o processo filho termine antes do usuário emitir o comando 't' o processo controlador deve notificar o término ao usuário e reiniciar o ciclo de controle.

Chamadas de sistema envolvidas em C++: `vfork`, `signal` e `kill`

Módulos.funções envolvidas em Python: `subprocess.Popen`, `signal.signal` e `os.kill`

# Programa

Inicialmente, escrevemos o programa que será o processo submetido aos comandos de parada, continuação e finalização.

## ```contagem.py```



````python
import time
import sys

i = 0

while i < 30:
	if i == 0:
		print(i, end = '', flush = True)
		time.sleep(0.5)
		i = i + 1
	else:
		print(" ", i, end = '', flush = True)
		time.sleep(0.5)
		i = i + 1
sys.exit()
````

A partir disso, podemos, então, escrever nosso programa que será capaz de controlar o processo de ```contagem.py```.

## ```controla_processos.py```

````python
import subprocess
import signal
import os
import time
#python3 asteriscos.py

aux = True                                     #variavel de loop while

args = input("linha de comando no terminal para executar o programa desejado:\n ")
args = args.split()                            #damos um split na entrada para separar corretamente os comandos

prog = subprocess.Popen(args)                  #lendo o programa do processo filho

while aux:
	if prog.pid != 0:                            #se não for o processo filho, isto é, se for o processo pai, criamos um menu explicativo dos poossíveis inputs
		print("Digite um dos comandos abaixo:\n- s: suspende a execução do processo filho\n- c: continua a execução do processo filho\n- t: termina a execução do processo filho")
		aux = input()                              #pegando inputs do teclado
		if aux == 's':                             #se o input é 's' para parar
			if prog.poll() != None:                  #checando se o prpograma já terminou
				                                       #se sim, inciamos de novo
				print("Programa já terminou, mas vamos iniciá-lo novamente\n-----------------------------------------------------------\nPrograma reinicializado")
				prog = subprocess.Popen(args)          #reinicializando o processo filho
			else:                                    #se não, continuamos normalmente com a ação de suspensão
				prog.send_signal(signal.SIGSTOP)       #fazendo o processo filho parar
				print("Execução do processo filho suspensa.\n-----------------------------------------------------------")
		if aux == 'c':                             #se o input é 'c' para continuar
			if prog.poll() != None:                  #checando se o prpograma já terminou
				                                       #se sim, inciamos de novo
				print("Programa já terminou, mas vamos iniciá-lo novamente\n-----------------------------------------------------------\nPrograma reinicializado")
				prog = subprocess.Popen(args)          #reinicializando o processo filho
			else:                                    #se não, continuamos normalmente com a a ação de continuação
				prog.send_signal(signal.SIGCONT)       #fazendo processo filho continuar
				print("Execução do processo filho continuada.\n-----------------------------------------------------------")
		if aux == 't':                             #se o input é 't' para terminar
			prog.kill()                              #matando a execução por completa do programa
			print("Processo filho terminado.\n-----------------------------------------------------------")
			aux = False                              #terminamos o loop while
````
