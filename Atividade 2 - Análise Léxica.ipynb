Como exigido no enunciado, uma estrutura de pilha deve ser utilizada para esta atividade. 

Assim, para não ficar restrito ao uso de estruturas nativas da linguagem, vamos criar uma estrutura de pilha através de classes.

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

#Criação da Estrutura de Pilha
#Como exigido no enunciado, uma estrutura de pilha deve ser utilizada para 
#esta atividade. Assim, para não ficar restrito ao uso de estruturas nativas 
#da linguagem, vamos criar uma estrutura de pilha através de classes.

class Stack:
  def __init__(self):
      self.items = []

  def isEmpty(self):
      return self.items == []

  def push(self, item):
      self.items.append(item)

  def pop(self):
      return self.items.pop()

  def peek(self):
      return self.items[len(self.items)-1]

  def size(self):
      return len(self.items)
  
#########################################################
#bibliotecas
import sys
sys.path.append("../..")
import ply.lex as lex

#tokens
tokens = ['NUMBER', 'PLUS', 'MINUS', 'DIV', 'MULT', 'CLEAR', 'SIGNAL']

t_PLUS = r'\+'

t_MINUS = r'\-'

t_DIV = r'\/'

t_MULT = r'\*'

t_CLEAR = r'C'

t_SIGNAL = r'S'

#definindo todos os números de entrada como float
def t_NUMBER(t):
  r'[-]?[0-9]+'
  t.value = float(t.value)
  return t

#inicializando a pilha
pilha = Stack()

#programa da calculadora
lexer = lex.lex()

aux = True

if __name__ == "__main__":
    while aux == True:
        lexer.input(input(''))                           #lendo o input do teclado
        tok = lexer.token()                              #lendo o token inserido
        
        if tok == None:                                  #se acabou os dados
            print("Acabaram-se os dados para se realizar operações.")
            aux = False
        
        elif tok.type == 'NUMBER':                       #se for um número inserido
            pilha.push(tok.value)                        #número é inserido na pilha
            print("Situação da Pilha (topo é o elemento mais a direita):", pilha.items)
        
        elif tok.type == 'PLUS':                         #se for o sinal de +
            n1 = pilha.pop()                             #remove o primeiro número
            n2 = pilha.pop()                             #remove o segundo número
            pilha.push(n1+n2)                            #adiciona o resultado na pilha
            print(n1+n2)
            print("Situação da Pilha (topo é o elemento mais a direita):", pilha.items)
        
        elif tok.type == 'MINUS':                        #se for o sinal de -
            n1 = pilha.pop()                             #remove o primeiro número
            n2 = pilha.pop()                             #remove o segundo número
            pilha.push(n1-n2)                            #adiciona o resultado na pilha
            print(n1-n2)
            print("Situação da Pilha (topo é o elemento mais a direita):", pilha.items)
        
        elif tok.type == 'DIV':                          #se for o sinal de /
            n1 = pilha.pop()                             #remove o primeiro número
            n2 = pilha.pop()                             #remove o segundo número
            if n1 == 0:
                print("Math ERROR: Divisão por zero é inválida!")
                #limpamos os dados da calculadora
                while pilha.isEmpty() == False:          #enquanto a pila estiver cheia
                    pilha.pop()                          #removemos os elementos existentes
                print("Situação da Pilha (topo é o elemento mais a direita):", pilha.items)
            else:
                    pilha.push(n1/n2)                    #adiciona o resultado na pilha
                    print(n1/n2)
                    print("Situação da Pilha (topo é o elemento mais a direita):", pilha.items)
        
        elif tok.type == 'MULT':                         #se for o sinal de *
            n1 = pilha.pop()                             #remove o primeiro número
            n2 = pilha.pop()                             #remove o segundo número
            pilha.push(n1*n2)                            #adiciona o resultado na pilha
            print(n1*n2)
            print("Situação da Pilha (topo é o elemento mais a direita):", pilha.items)
        
        elif tok.type == 'CLEAR':                        #se for o sinal para limpar a calculadora
            while pilha.isEmpty() == False:              #enquanto a pila estiver cheia
                pilha.pop()                              #removemos os elementos existentes
            print("Situação da Pilha (topo é o elemento mais a direita):", pilha.items)
        
        elif tok.type == 'SIGNAL':                       #se for para inversão de sinal do topo
            n1 = pilha.pop()                             #removemos o primeiro elemento
            pilha.push(-n1)                              #adicionamos o elemento com sinal trocado
            print("Situação da Pilha (topo é o elemento mais a direita):", pilha.items)
```
